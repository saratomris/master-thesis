Which suggests inversion symmetry about $\mathbf{q}$ of our unbarred effective potentials.

Why is this so?


We must study the components of $V^{\sigma_1\sigma_2\sigma_3\sigma_4}_{\mathbf{k}\mathbf{k}'}$ on the Fermi surface:

$$
\begin{align}
\left. V^{\sigma_1\sigma_2\sigma_3\sigma_4}_{\mathbf{k}\mathbf{k}'} \right|_{FS} 
    &= -\frac12 \sum_{\gamma} \frac{1}{\omega_{\mathbf{q}\gamma}} \mathcal{A}^{\sigma_1\sigma_2\sigma_3\sigma_4}_{\mathbf{q}\gamma\nu} \\
    &= -\frac12 \sum_{\gamma} \frac{1}{\omega_{\mathbf{q}\gamma}} \left( - A^{\sigma_1\sigma_4\sigma_2\sigma_3}_{\mathbf{q}\gamma\nu\bar{\nu}} - A^{\sigma_2\sigma_3\sigma_1\sigma_4}_{-\mathbf{q}\gamma\bar{\nu}\nu} \right) \\
    &\begin{aligned}
        = \frac12 \sum_{\gamma,L,L'} \frac{1}{\omega_{\mathbf{q}\gamma}} 
        \Big(
            &g_{L\nu\sigma_1\sigma_4}g_{L'\bar{\nu} \sigma_2\sigma_3}u_{\mathbf{q} L\gamma}v_{-\mathbf{q} L'\gamma} +g_{L\nu\sigma_1\sigma_4}g^\ast_{L'\nu\sigma_3\sigma_2}u_{\mathbf{q} L\gamma}u^\ast_{\mathbf{q} L'\gamma}
            +g^\ast_{L\bar{\nu}\sigma_4\sigma_1}g_{L'\bar{\nu}\sigma_2\sigma_3}v^\ast_{-\mathbf{q} L\gamma}v_{-\mathbf{q} L'\gamma} +g^\ast_{L\bar{\nu}\sigma_4\sigma_1}g^\ast_{L'\nu \sigma_3\sigma_2}v^\ast_{-\mathbf{q} L\gamma}u^\ast_{\mathbf{q} L'\gamma}
            \\
            &+g_{L\bar{\nu}\sigma_2\sigma_3}g_{L'\nu \sigma_1\sigma_4}u_{-\mathbf{q} L\gamma}v_{\mathbf{q} L'\gamma} +g_{L\bar{\nu}\sigma_2\sigma_3}g^\ast_{L'\bar{\nu}\sigma_4\sigma_1}u_{-\mathbf{q} L\gamma}u^\ast_{-\mathbf{q} L'\gamma}
            +g^\ast_{L\nu\sigma_3\sigma_2}g_{L'\nu\sigma_1\sigma_4}v^\ast_{\mathbf{q} L\gamma}v_{\mathbf{q} L'\gamma} +g^\ast_{L\nu\sigma_3\sigma_2}g^\ast_{L'\bar{\nu} \sigma_4\sigma_1}v^\ast_{\mathbf{q} L\gamma}u^\ast_{-\mathbf{q} L'\gamma}
        \Big)
     \end{aligned}
\end{align}
$$

First, we study this for various $L$.

### Plott: Veff(L)


```python
## Functions 

def Aq_elements_L(L, s1,s2,s3,s4, gamma,Tinv_m_,  Qnu_array_,  period=n, nrows=n_chains, disable_tqdm=False):
    A = np.zeros(Tinv_m_.shape[0], dtype=np.complex_)

    for Lp in range(m):
        A +=  gLnsts(L,Qnu_array_,s1,s2, period,nrows)*gLnsts(Lp,-Qnu_array_,s3,s4, period,nrows)\
                * Tinv_m_[:,L,gamma%m]* np.conjugate(Tinv_m_[:,Lp+m,gamma%m])\
                + gLnsts(L,Qnu_array_,s1,s2, period,nrows)*np.conjugate(gLnsts(Lp,Qnu_array_,s4,s3, period,nrows))\
                * Tinv_m_[:,L,gamma%m]* np.conjugate(Tinv_m_[:,Lp,gamma%m])\
                + np.conjugate(gLnsts(L,-Qnu_array_,s2,s1, period,nrows))*gLnsts(Lp,-Qnu_array_,s3,s4, period,nrows)\
                * Tinv_m_[:,L+m,gamma%m]* np.conjugate(Tinv_m_[:,Lp+m,gamma%m])\
                + np.conjugate(gLnsts(L,-Qnu_array_,s2,s1, period,nrows))*np.conjugate(gLnsts(Lp,Qnu_array_,s4,s3, period,nrows))\
                * Tinv_m_[:,L+m,gamma%m]* np.conjugate(Tinv_m_[:,Lp,gamma%m])
    return A


def curlyAq_elements_L(L, s1,s2,s3,s4, gamma,Tinv_m_p_, Tinv_m_n_,  Qnu_array_,  period=n, nrows=n_chains, disable_tqdm=False):
    return - Aq_elements_L(L, s1,s4,s2,s3, gamma,Tinv_m_p_,  Qnu_array_,  period, nrows, disable_tqdm) \
           - Aq_elements_L(L, s2,s3,s1,s4, gamma,Tinv_m_n_, -Qnu_array_,  period, nrows, disable_tqdm)

def get_Veff_L(L, s1,s2,s3,s4, k_s,kp_s):
    Veff_L = 0
    q_FS, Qnu_array = get_q_Qnu_FS(None, None,\
                                th_array_, th0i=th0i_, eta=eta, disable_tqdm=True, k_sign_=k_s, kp_sign_=kp_s)
    omegas_, Tinv_p_ = get_2m_bands(q_FS, None, None, disable_tqdm_qs=True, k_sign_=k_s, kp_sign_=kp_s)

    Tinv_n_ = np.zeros_like(Tinv_p_)
    Tinv_n_[:, :m,:m] = Tinv_p_[:, m:,m:].conj()
    # Tinv_n_[:, :m,m:] = Tinv_p_[:, m:,:m].conj()
    Tinv_n_[:, m:,:m] = Tinv_p_[:, :m,m:].conj()
    # Tinv_n_[:, m:,m:] = Tinv_p_[:, :m,:m].conj()

    for gamma in range(m):
        Veff_L += -1/2* curlyAq_elements_L(L, s1,s2,s3,s4, gamma,Tinv_p_[:,:,:m], Tinv_n_[:,:,:m],  Qnu_array,  period=n, nrows=n_chains, disable_tqdm=True)/omegas_[:,gamma]

    return Veff_L


def plot_Veff_L(s1,s2,s3,s4, k_s,kp_s):
    fig_L, axes_L = plt.subplots(1,2, figsize=(8,4))
    axes_L = axes_L.flatten()
    sta = { -1: u"\u2193", 1 : u"\u2191" }   ## spin_to_arrows dict
    sts = { -1: "-", +1: "+"}                ## sign to string dict
    plt.suptitle(r"$N  V^{{{}{}{}{}}}_{{({}k)({}k')}} $".format(sta.get(s1), sta.get(s2), sta.get(s3), sta.get(s4), sts.get(k_s),sts.get(kp_s)) + r" for various $L$")

    all_lines_L = []
    annotations_L = []
    annotations_coords_L = []
    real_imag_str = ["Real", "Imag"]
    real_imag_func = lambda x : [np.real(x), np.imag(x)]
    for idx,ax_L in enumerate(axes_L):
        lines_L=[]
        ax_L.set_title(real_imag_str[idx] + " part")
        ax_L.set_xticks(theta_tick_values,   theta_tick_labels)
        ax_L.set_xlabel(r"$\theta$")
        ax_L.xaxis.set_minor_locator(MultipleLocator(np.pi/4))
        ax_L.set_xlim(th_array_[0],th_array_[-1])

        ax_L.tick_params(axis='y', labelsize=10)
        ax_L.grid(c='gainsboro', which='both')
        for L in range(m):
            tmp = get_Veff_L(L, s1,s2,s3,s4, k_s,kp_s)
            l, = ax_L.plot(th_array_, real_imag_func(tmp)[idx], label=fr"$L={L}$", dashes=(5,0+10*bool(L==7)*bool(idx==0)))
            lines_L.append(l)
        all_lines_L.append(lines_L)
        annot_L = ax_L.annotate("", xy=(0, 0), xytext=(-20, 20), textcoords="offset points",
                            bbox=dict(boxstyle="round", fc="w"),
                            arrowprops=dict(arrowstyle="->"))
        annot_L.set_visible(False)
        annot_coords_L = ax_L.annotate("", xy=(0, 0), xytext=(-20, -30), textcoords="offset points",
                            bbox=dict(boxstyle="round", fc="w"),
                            arrowprops=dict(arrowstyle="->"))
        annot_coords_L.set_visible(False)
        annotations_L.append(annot_L)
        annotations_coords_L.append(annot_coords_L)
    
    plt.tight_layout()

    h_multiple_L = lambda evnt : hover_multiple(evnt, fig_L,axes_L,annotations_L,all_lines_L)
    oc_t_multiple_L = lambda evnt : onclick_theta_multiple(evnt, annotations_coords_L,all_lines_L)
    fig_L.canvas.mpl_connect("motion_notify_event", h_multiple_L)
    fig_L.canvas.mpl_connect('button_press_event', oc_t_multiple_L)
    plt.show()
```

Which indicates that there exists sets $(L,\bar{L})$ such that all unbarred effective potentials of interest cancel.
Moreover, all these sets are equal for the effective potentials in consideration.

We seek to understand why.

Could this be due to the sum over $\bar{L}$, or is there a set $(\bar{L}, \bar{L}')$ for each $(L,L')$ such that $V^{\uparrow\uparrow\uparrow\downarrow}_{\mathbf{k}\mathbf{k}'}(L, L') = - V^{\uparrow\uparrow\uparrow\downarrow}_{\mathbf{k}\mathbf{k}'}(\bar{L}, \bar{L}')$ (and accordingly for each term in $V^{\uparrow\uparrow\uparrow\downarrow}_{\mathbf{k}\mathbf{k}'}(L, L')$)?


Checking $V^{\uparrow\uparrow\uparrow\downarrow}_{\mathbf{k}\mathbf{k}'}(L, L')$ for various $L, L'$:

```python
## Functions

def Aq_elements_L_Lp(L,Lp, s1,s2,s3,s4, gamma,Tinv_m_,  Qnu_array_,  period=n, nrows=n_chains, disable_tqdm=False):
    m = Tinv_m_.shape[-1]

    return  gLnsts(L,Qnu_array_,s1,s2, period,nrows)*gLnsts(Lp,-Qnu_array_,s3,s4, period,nrows)\
            * Tinv_m_[:,L,gamma%m]* np.conjugate(Tinv_m_[:,Lp+m,gamma%m])\
            + gLnsts(L,Qnu_array_,s1,s2, period,nrows)*np.conjugate(gLnsts(Lp,Qnu_array_,s4,s3, period,nrows))\
            * Tinv_m_[:,L,gamma%m]* np.conjugate(Tinv_m_[:,Lp,gamma%m])\
            + np.conjugate(gLnsts(L,-Qnu_array_,s2,s1, period,nrows))*gLnsts(Lp,-Qnu_array_,s3,s4, period,nrows)\
            * Tinv_m_[:,L+m,gamma%m]* np.conjugate(Tinv_m_[:,Lp+m,gamma%m])\
            + np.conjugate(gLnsts(L,-Qnu_array_,s2,s1, period,nrows))*np.conjugate(gLnsts(Lp,Qnu_array_,s4,s3, period,nrows))\
            * Tinv_m_[:,L+m,gamma%m]* np.conjugate(Tinv_m_[:,Lp,gamma%m])


def curlyAq_elements_L_Lp(L,Lp, s1,s2,s3,s4, gamma,Tinv_m_p_, Tinv_m_n_,  Qnu_array_,  period=n, nrows=n_chains, disable_tqdm=False):
    return - Aq_elements_L_Lp(L,Lp, s1,s4,s2,s3, gamma,Tinv_m_p_,  Qnu_array_,  period, nrows, disable_tqdm) \
           - Aq_elements_L_Lp(L,Lp, s2,s3,s1,s4, gamma,Tinv_m_n_, -Qnu_array_,  period, nrows, disable_tqdm)

def get_Veff_L_Lp(L,Lp, s1,s2,s3,s4, Tinv_m_p_,Tinv_m_n_, Qnu_array_, omegas_):
    Veff_L_Lp = 0
    for gamma in range(m):
        Veff_L_Lp += -1/2* curlyAq_elements_L_Lp(L,Lp, s1,s2,s3,s4, gamma,Tinv_m_p_, Tinv_m_n_,  Qnu_array_,  period=n, nrows=n_chains, disable_tqdm=True)/omegas_[:,gamma]
    return Veff_L_Lp


def get_L_Lp_Lb_Lbp_set_and_L_Lp_null_set__Veff_L_Lp(s1,s2,s3,s4, k_s,kp_s,  q_idx=0, tol=1e-12):
    Veff_L_Lp_array = np.zeros((m,m, Nth), dtype=np.complex_)

    q_FS_, Qnu_array_ = get_q_Qnu_FS(None, None,\
                                th_array_, th0i=th0i_, eta=eta, disable_tqdm=True, k_sign_=k_s, kp_sign_=kp_s)
    omegas_, Tinv_p_ = get_2m_bands(q_FS_, None, None, disable_tqdm_qs=True, k_sign_=k_s, kp_sign_=kp_s)

    Tinv_n_ = np.zeros_like(Tinv_p_)
    Tinv_n_[:, :m,:m] = Tinv_p_[:, m:,m:].conj()
    # Tinv_n_[:, :m,m:] = Tinv_p_[:, m:,:m].conj()
    Tinv_n_[:, m:,:m] = Tinv_p_[:, :m,m:].conj()
    # Tinv_n_[:, m:,m:] = Tinv_p_[:, :m,:m].conj()

    for L in range(m):
        for Lp in range(m):
            Veff_L_Lp_array[L,Lp,:] = get_Veff_L_Lp(L,Lp, s1,s2,s3,s4, Tinv_p_[:,:,:m],Tinv_n_[:,:,:m], Qnu_array_, omegas_)

    Veff_L_Lp_q = Veff_L_Lp_array[:,:,q_idx]

    L_Lb_set, null_set_L = get_L_Lb_set_and_Lnull_set__Veff_L(s1,s2,s4,s3, k_s,kp_s)

    L_Lp_Lb_Lbp_set = []
    null_set_L_Lp   = []
    for L, Lb in sorted(L_Lb_set + [(nL,nL) for nL in null_set_L]):
        for Lp in range(m):
            for idx in np.argwhere(np.abs(((Veff_L_Lp_q[Lb, :] + Veff_L_Lp_q[L,Lp])/(Veff_L_Lp_q[Lb, :] - Veff_L_Lp_q[L,Lp])).real) < tol).flatten():
                L_Lp_Lb_Lbp_set += [[(L,Lp), (Lb, idx)]]
            if np.abs(Veff_L_Lp_q[L,Lp].real) < tol:
                null_set_L_Lp += [(L,Lp)]
    return L_Lp_Lb_Lbp_set, null_set_L_Lp
```

We proceed to consider the symmetries of our lattice in the sought to understand why our system yields the above relations.
Due to the nature of our spiral-phase spin structure, there will always exists a sub-lattice $\bar{L}$ for each $L$ such that $g_{L\nu\sigma_1\sigma_2} = -g_{\bar{L}\nu\sigma_2\sigma_1}$.
However, for the relations (1)-(8) to hold, these functions in product with the Bogoliubov transformation matrix elements must equate.

We conclude that any deeper insight into the absence of strong topological superconductivity in our system is obtained by studying the analytical expressions of the elements $u$ and $v$ of the Bogoliubov transformation matrix.
